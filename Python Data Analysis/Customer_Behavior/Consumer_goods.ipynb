# Loading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data into the dataset

df = pd.read_excel('Consumer Good Sales for PandasAI Case Study.xlsx')
```

```python
df.info()
```

```python
df.columns
```

```python
df[[ 'DeliveryStreetAddress', 'DeliverySuburb', 'DeliveryCity',
       'DeliveryRegion', 'DeliveryCountry', 'DeliveryPostCode',
       'DeliveryMethod','Line Comments']] = df[[ 'DeliveryStreetAddress', 'DeliverySuburb', 'DeliveryCity',
       'DeliveryRegion', 'DeliveryCountry', 'DeliveryPostCode',
       'DeliveryMethod','Line Comments']].astype(str)
df.info()
```

```python


# formatting the Unit Price and Line Total float values to 2 decimals
pd.options.display.float_format = '{:.2f}'.format

# Replacing empty or null values with blank
df.replace('nan','')
df.head(5)
```

```python
# Dropping any duplicate Order Numbers
df.drop_duplicates('OrderNumber')
```

```python
# Counting unique values
df.nunique()
```

```python
# Creating Month-Year Column to identify and group sales for the particular month
df['Month-Year'] = df['OrderDate'].dt.strftime('%b-%y')
df['Month_Name'] = df['OrderDate'].dt.strftime('%b')
df['Years'] = df['OrderDate'].dt.strftime('%y')
df.head(5)
```

```python
# Identifying Min and Max Year of the Sales
#df['Years'].max() =16
df['Years'].min() # = 15
```

```python
# Spliting into separate dataframes based on Year of Sale 
df_2015 = df[df["Years"] == "15"]
df_2016 = df[df["Years"] == "16"]
```

# Visualizing Sales for the Year 2015

```python
# df_Monthly_Sales['Month-Year','Total Count'] = df.groupby('Month-Year')['LineTotal'].count()
df2015 = df_2015.groupby('Month_Name')['LineTotal'].count().reset_index(name='Counts').sort_values(by='Month_Name', ascending=False)

#plt.figure(figsize= (20,7))
#plt.xlabel("Months")
#plt.ylabel("No. of Sales")
#plt.bar( df2['Month_Name'], df2['Counts']) 
#plt.show() 
#df2

# for displaying data labels over the bar
x = np.arange(len(df2015['Month_Name'])) # the label locations
width = 0.35 # the width of the bars

fig, ax = plt.subplots(figsize= (10,5))

ax.set_xlabel('Months')
ax.set_ylabel('No. of Sales')
ax.set_title('Warehouse Sales Data for Year 2015')
ax.set_xticks(x)
ax.set_xticklabels(df2015['Month_Name'])

pps = ax.bar(x - width/2, round(df2015['Counts'],1), width, label=df2015['Counts'])
for p in pps:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()
```

# Visualizing Sales for the Year 2016

```python
# df_Monthly_Sales['Month-Year','Total Count'] = df.groupby('Month-Year')['LineTotal'].count()
df2016 = df_2016.groupby('Month_Name')['LineTotal'].count().reset_index(name='Counts').sort_values(by='Month_Name', ascending=False)

# for displaying data labels over the bar
x = np.arange(len(df2016['Month_Name'])) # the label locations
width = 0.35 # the width of the bars

fig, ax = plt.subplots(figsize= (10,5))

ax.set_xlabel('Months')
ax.set_ylabel('No. of Sales')
ax.set_title('Warehouse Sales Data for Year 2016')
ax.set_xticks(x)
ax.set_xticklabels(df2016['Month_Name'])

pps = ax.bar(x - width/2, round(df2016['Counts'],1), width, label=df2016['Counts'])
for p in pps:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()
```

# Categorizing Sales By Warehouse Type

```python
# Categorizing Total sales based on Warehouse Code
df3 = df.groupby('WarehouseCode')['LineTotal'].sum().reset_index(name='Total Sales').sort_values(by='Total Sales', ascending=False)
# df['ProductCode']
#plt.figure(figsize= (10,5))
plt.ylim(500000,7000000)
plt.title("Warehouse Sales Data")
plt.xlabel("Warehouse")
plt.ylabel("Total Sales (Millions)")

plt.bar( df3['WarehouseCode'], df3['Total Sales']) 
plt.show() 
df3
```

```python
x = np.arange(len(df3['WarehouseCode'])) # the label locations
width = 0.35 # the width of the bars

fig, ax = plt.subplots()

ax.set_ylabel('Total Sales (in million)')
ax.set_title('Warehouse Sales Data')
ax.set_xticks(x)
ax.set_xticklabels(df3['WarehouseCode'])
ax.set_ylim(500000,7000000)

pps = ax.bar(x - width/2, round(df3['Total Sales'],1), width, label=df3['Total Sales'])
for p in pps:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()
```
